```python
mac_tcs,time_tcs,time_mean_tcs=[],[],[]
for i in set(df_tcs.mac):
    s=df_tcs[df_tcs['mac']==i]
    if len(s)!=1:
        mac_tcs.append(i)
        time_tcs.append(s.Timestamp.max()-s.Timestamp.min())
        time_mean_tcs.append(s.Strength.mean())
```

```python
t1=pd.DataFrame(mac_tcs,columns=['mac'])
t2=pd.DataFrame(time_tcs,columns=['time'])
t3=pd.DataFrame(time_mean_tcs,columns=['strength'])
```

```python
t_tcs = pd.concat([t1, t2,t3], axis=1).reindex(t1.index)
```

```python
mac_vij,time_vij,str_vij=[],[],[]
for i in set(df_vij.mac):
    k=df_vij[df_vij['mac']==i]
    if len(k)!=1:
        mac_vij.append(i)
        time_vij.append(k.Timestamp.max()-k.Timestamp.min())
        str_vij.append(k.Strength.mean())
```

```python
u1=pd.DataFrame(mac_vij,columns=['mac'])
u2=pd.DataFrame(time_vij,columns=['time'])
u3=pd.DataFrame(str_vij,columns=['strength'])
```

```python
t_vij = pd.concat([u1, u2,u3], axis=1).reindex(u1.index)
```
